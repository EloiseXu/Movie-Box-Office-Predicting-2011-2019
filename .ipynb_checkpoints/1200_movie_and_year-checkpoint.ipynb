```python
from lxml import etree
import requests
import pandas as pd
```

```python
results = []

for year in range(2011, 2020):
    input_file = 'movies/movies-%s.txt' % str(year)
    fobj = open(input_file, 'r')
    content = fobj.readlines()
    for i in range(len(content)):
        content[i] = content[i].strip()
        results.append([content[i], year])
    fobj.close()
```

```python
df = pd.DataFrame(results, columns=['name', 'year']) 
```
