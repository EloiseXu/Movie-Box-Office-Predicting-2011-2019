```python
def exec_spider(name, year, month, day):
    global browser
    try:
        browser.find_elements_by_xpath("//form[@id='search-input-form']/input[3]")[0].clear()
        browser.find_elements_by_xpath("//form[@id='search-input-form']/input[3]")[0].send_keys(name)
        try:
            browser.find_element_by_xpath("//*[@id='home']/div[2]/div[2]/div/div[1]/div/div[2]/div/span/span").click()
        except:
            try:
                browser.find_element_by_xpath("/html/body/div[1]/div[1]/div/div[2]/div/div[1]/div/span/span").click()
            except:
                return False
        time.sleep(2)

        fyear,fmonth,ayear,amonth=CalculateDate(year,month)
        browser.maximize_window()
        
        halfy = get_index(fyear, fmonth, ayear, amonth, 2019, 2019, 5, 6)
        if halfy is False:
            return False
        
        browser.refresh()
        time.sleep(10)
        
        halfago = get_index(ayear, amonth, ayear, amonth, 2019, 2019, 5, 6)
        if halfago is False:
            return False
        
        browser.refresh()
        time.sleep(10)
        
        release = get_index(fyear, fmonth, fyear, fmonth, 2019, 2019, 5, 6)
        if release is False:
            return False
        
        halfy = int(re.sub("[\!\%\[\]\，\。\·\：\、\!\t\,]", "", halfy))
        halfago = int(re.sub("[\!\%\[\]\，\。\·\：\、\!\t\,]", "", halfago))
        release = int(re.sub("[\!\%\[\]\，\。\·\：\、\!\t\,]", "", release))
        
        return halfy, release-halfago
    except:
        return False
```

```python
def CalculateDate(year,month):
    fdate = datetime.date(int(year), int(month), 15)
    adate = fdate - relativedelta(months=6)
    
    fyear = int(year)
    fmonth = int(month)
    ayear = int(adate.year)
    amonth = int(adate.month)
    
    if ayear == 2010:
        ayear = 2011
        amonth = 1
    
    return fyear,fmonth,ayear,amonth
```
